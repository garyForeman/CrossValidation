#Data Generation

Here I generate data for my cross validation example. The dependent variable is generated randomly from a tenth order Legendre polynomial with Gaussian perterbations. The independent variable is drawn from the interval [-1,1).

##Imports

```python
%matplotlib inline

import matplotlib.pyplot as plt
import numpy as np

from sklearn.preprocessing import PolynomialFeatures
```

##Simulated Function

I use the P<sub>10</sub>(x) Legendre polynomial to ensure the underlying function requires a decent number of parameters to model properly. Further, P<sub>10</sub> is ideal as a tenth order polynomial because all eight of its inflection points lie within the interval [-1,1). Let's see what the function looks like.

```python
x_legendre = np.linspace(-1, 1, 201).reshape(201, 1)
coef = np.array([0]*10 + [1])
P_10 = np.polynomial.legendre.Legendre(coef)
plt.plot(x_legendre, P_10(x_legendre), 'r')
plt.ylabel(r'$P_{10}(x)$', size=15)
```

##Simulated data

For the purposes of our example, we'll introduce random Gaussian perturbations with zero mean and standard deviation of 0.1. Below I'll generate the data, plot it, then write it to a separate file, so that it may be loaded into the modeling notebook.

```python
x = np.random.rand(250) * 2. - 1.
y = P_10(x) + np.random.randn(len(x)) * 0.1
plt.plot(x_legendre, P_10(x_legendre), 'r')
plt.plot(x, y, '.')
np.save('x_arr', x)
np.save('y_arr', y)
```
